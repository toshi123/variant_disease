# humsavarにあるACだけのac2go.jsonをつくる

```python
import json

# ファイルパスの設定
humsavar_file_path = "./humsavar_20240604.json"  # humsavar.jsonのパス
ac2go_file_path = "./ac2go.json"        # ac2go.jsonのパス
output_file_path = "./ac2go_humsavar.json"  # 出力先のパス

# humsavar.jsonの読み込み
with open(humsavar_file_path, "r") as f:
    humsavar_data = json.load(f)

# humsavar.jsonに含まれるUniProt_ACを取得
humsavar_ac_list = set()
for ac in humsavar_data:
    humsavar_ac_list.add(ac)

# ac2go.jsonの読み込み
with open(ac2go_file_path, "r") as f:
    ac2go_data = json.load(f)

# humsavarのACに該当するエントリを抽出
filtered_ac2go_data = {ac: go_data for ac, go_data in ac2go_data.items() if ac in humsavar_ac_list}

# 新しいJSONファイルとして保存
with open(output_file_path, "w") as f:
    json.dump(filtered_ac2go_data, f, indent=4)

print(f"Filtered data saved to {output_file_path}")
```
